Em relação aos óbitos, o ES sofreu em grande amplitude com a que começou por volta do mês de março de 2021. Atingindo uma média móvel de casos de mais de 80 óbitos/dia.
O número de mortes começou a cair no final de abril no início de maio, possivelmente influenciadas pela vacinação e por medidas mais rígidas de isolamento social. 

Em relação aos municípios, São José do Calçado ocupa a liderança da relação entre o total de óbitos e a população infectada com cerca de 6% dos doentes morrendo de COVID-19

É interessante notar que a cidade de São José do Calçado ocupa as últimas posições entre os municípios com mais casos. A doença é mais letal nessa cidade? O mesmo acontece na Grande Vitória. A cidade de Cariacica ocupa os últimos lugares referentes a ocorrência da doença mas figura entre os primeiros na taxa de óbitos. 

De fato, não existe nenhuma explicação biológica para que algumas cidades tão próximas tenham tão diferentes e elevados valores de mortalidade. Sendo assim, duas possibilidades são mais factíveis. A primeira, a subnotificação do número de casos. Ou seja, essas cidades falham em rastrear todos os casos ativos de COVID-19. 

Em 100 pessoas infectadas os esperado é que aproximadamente duas morram (2%), mas se mesmo tendo 100 pessoas infectadas a cidade só consegue detectar 50 pessoas com o vírus e de novo, moram duas pessoas a taxa de óbitos sobe para 4%.

O segundo ponto, diz respeito ao acesso a assistência de saúde, cidades mais pobres tendem a ter cidadãos com menos acesso à saúde de qualidade  o que pode levar ao desenvolvimento de mais pessoas com comorbidades negligenciada, elevando assim o risco de desenvolver quadros graves de COVID. 


Vamos observar essas hipóteses observando como cada estado da região Sudeste se comporta na pandemia.
